```python
import pandas as pd
```

```python
kospi_real = pd.read_csv('../data/코스피실제주가_1112_1119.csv')
kospi_estimate = pd.read_csv('../data/코스피예측주가_1011_1120.csv')
kosdaq_real = pd.read_csv('../data/코스닥실제주가_1112_1119.csv')
kosdaq_estimate = pd.read_csv('../data/코스닥예측주가_1011_1120.csv')
```

```python
# YYYY-MM-DD 형식으로 변환
kospi_real['Date'] = pd.to_datetime(kospi_real['Date']).dt.strftime('%Y-%m-%d')
kosdaq_real['Date'] = pd.to_datetime(kosdaq_real['Date']).dt.strftime('%Y-%m-%d')
```

```python
# 행렬 전환
kospi_real_T = pd.DataFrame(kospi_real.T.reset_index().iloc[1:].values, columns=kospi_real.T.reset_index().iloc[0].values)
kospi_estimate_T = pd.DataFrame(kospi_estimate.T.reset_index().iloc[1:].values, columns=kospi_estimate.T.reset_index().iloc[0].values)
kosdaq_real_T = pd.DataFrame(kosdaq_real.T.reset_index().iloc[1:].values, columns=kosdaq_real.T.reset_index().iloc[0].values)
kosdaq_estimate_T = pd.DataFrame(kosdaq_estimate.T.reset_index().iloc[1:].values, columns=kosdaq_estimate.T.reset_index().iloc[0].values)
```

```python
# 'Date' 컬럼 이름을 'Ticker'로 변경
kospi_real_T.columns = ['Ticker'] + list(kospi_real_T.columns[1:])
kospi_estimate_T.columns = ['Ticker'] + list(kospi_estimate_T.columns[1:])
kosdaq_real_T.columns = ['Ticker'] + list(kosdaq_real_T.columns[1:])
kosdaq_estimate_T.columns = ['Ticker'] + list(kosdaq_estimate_T.columns[1:])
```

```python
# 'Ticker' 기준으로 오름차순 정렬
kospi_estimate_T = kospi_estimate_T.sort_values(by='Ticker', ascending=True).reset_index(drop=True)
kosdaq_estimate_T = kosdaq_estimate_T.sort_values(by='Ticker', ascending=True).reset_index(drop=True)
```
